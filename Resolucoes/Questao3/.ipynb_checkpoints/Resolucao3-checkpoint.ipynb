### Passos para segmentação com FastAI:
- Instanciar ```SegmentationDataLoaders``` para treinar um modelo de segmentação.

```
dls = SegmentationDataLoaders.from_label_func(path,
                        bs=8,
                        fnames = get_image_files(path/"images"),
                        label_func = lambda o: path/'labels'/f'{o.stem}_P{o.suffix}',
                        codes = np.loadtxt(path/'codes.txt', dtype=str) )
```

- Para utilizar esta função, precisamos das imagens de satélite juntamente com as imagens da máscara de localização das construções.
- Podemos gerar as máscaras a partir dos arquivos de anotação JSON.
- Além disso, como mencionado anteriormente, não será necessário utilizar todo o dataset. Desta forma, podemos selecionar aleatoriamente um subconjunto para treinamento.


````python
from pycocotools.coco import COCO
from pycocotools import mask as cocomask

import cv2
import matplotlib.pyplot as plt

import random
import os

coco = COCO('annotation.json')
category_ids = coco.loadCats(coco.getCatIds())

# This generates a list of all `image_ids` available in the dataset
image_ids = coco.getImgIds(catIds=coco.getCatIds())

random_image_id = random.choice(image_ids)
img = coco.loadImgs(random_image_id)[0]

annotation_ids = coco.getAnnIds(imgIds=img['id'])
annotations = coco.loadAnns(annotation_ids)

ann_segmentation = annotations[0]['segmentation']

rle = cocomask.frPyObjects(ann_segmentation, img['height'], img['width'])

image_path = os.path.join("train/images", img["file_name"])
I = cv2.imread(image_path)
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
plt.imshow(I)
# Render annotations on top of the image
coco.showAnns(annotations)
plt.show()
````

````python
import numpy as np

mask_sum = np.zeros((img['height'], img['width']), dtype=np.uint8)  # Initialize with zeros

for ann in annotations:
    rle = cocomask.frPyObjects(ann['segmentation'], img['height'], img['width'])

    mask = cocomask.decode(rle)

    mask = np.asarray(mask, dtype=np.uint8) * 255
    
    mask_sum += mask[:,:,0]

plt.imshow(mask_sum, cmap='gray')  
plt.title("Segmentation Mask")
plt.axis('off')
plt.show()
````

````python
import shutil

num_train_images = 2
train_folder = 'my_train'

train_image_ids = random.sample(image_ids, num_train_images)

if not os.path.exists(train_folder):
    os.makedirs(train_folder)

for image_id in train_image_ids:
    img_info = coco.loadImgs(image_id)[0]
    print(img_info)
    # image_path = os.path.join('train/images', img_info['file_name'])
    # shutil.copy(image_path, os.path.join(train_folder, img_info['file_name']))

    annotation_ids = coco.getAnnIds(imgIds=img_info['id'])
    annotations = coco.loadAnns(annotation_ids)

    mask_sum = np.zeros((img_info['height'], img_info['width']), dtype=np.uint8)

    for ann in annotations:
        rle = cocomask.frPyObjects(ann['segmentation'], img_info['height'], img_info['width'])
    
        mask = cocomask.decode(rle)
    
        mask = np.asarray(mask, dtype=np.uint8) * 255
        
        mask_sum += mask[:,:,0]
        
    plt.imshow(mask_sum, cmap='gray')  
    plt.title("Segmentation Mask")
    plt.axis('off')
    plt.show()
````

````python
from pycocotools.coco import COCO
from pycocotools import mask as cocomask

import cv2
import matplotlib.pyplot as plt

import random
import os
import shutil
import numpy as np

num_train_images = 200
train_folder = 'my_segmentation2/my_train'

coco = COCO('annotation.json')
category_ids = coco.loadCats(coco.getCatIds())

# This generates a list of all `image_ids` available in the dataset
image_ids = coco.getImgIds(catIds=coco.getCatIds())

train_image_ids = random.sample(image_ids, num_train_images)

if not os.path.exists(train_folder):
    os.makedirs(train_folder)

for image_id in train_image_ids:
    img_info = coco.loadImgs(image_id)[0]

    image_path = os.path.join('train/images', img_info['file_name'])
    shutil.copy(image_path, os.path.join(train_folder, img_info['file_name']))

    annotation_ids = coco.getAnnIds(imgIds=img_info['id'])
    annotations = coco.loadAnns(annotation_ids)

    mask_sum = np.zeros((img_info['height'], img_info['width']), dtype=np.uint8)

    for ann in annotations:
        rle = cocomask.frPyObjects(ann['segmentation'], img_info['height'], img_info['width'])
    
        mask = cocomask.decode(rle)
    
        mask = np.asarray(mask, dtype=np.uint8) * 255
        
        mask_sum += mask[:,:,0]

    o = img_info['file_name']
    stem, suffix = o.split(".")
            
    cv2.imwrite(f'my_segmentation2/my_train_label/{stem}_P.{suffix}', np.clip(mask_sum, 0, 1))
````

````python
from fastbook import *
from fastai.vision.all import *
from pathlib import Path

path = Path("my_segmentation2")
dls = SegmentationDataLoaders.from_label_func(
    path, bs=8, fnames = get_image_files(path/"my_train"),
    label_func = lambda o: path/'my_train_label'/f'{o.stem}_P{o.suffix}',
    codes = np.loadtxt(path/'codes.txt', dtype=str))
learn = unet_learner(dls, resnet34)
learn.fine_tune(2)
````

````python
learn.show_results(max_n=2, figsize=(10,10))
````
